```python
import os
from dotenv import load_dotenv
import requests
```

```python
# Load environment variables
load_dotenv()
REC_API_KEY = os.getenv("RECREATION_GOV_API_KEY")

REC_API_URL = "https://ridb.recreation.gov/api/v1/facilities"

params = {
    "state": "UT",        # Example: Utah
    "facilityType": "Campground",
    "limit": 10
}

headers = {
    "apikey": REC_API_KEY
}

response = requests.get(REC_API_URL, params=params, headers=headers)

if response.status_code == 200:
    try:
        data = response.json()
        print(data)
    except Exception as e:
        print("Error decoding JSON:", e)
        print("Response text:", response.text)
else:
    print("Request failed with status code:", response.status_code)
    print("Response text:", response.text)
```
